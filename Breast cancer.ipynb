```python
df.drop(['Unnamed: 32', 'id'], axis = 1) 
print(df.shape) 
```

```python
def diagnosis_value(diagnosis): 
    if diagnosis == 'M': 
        return 1
    else: 
        return 0
  
df['diagnosis'] = df['diagnosis'].apply(diagnosis_value) 
```

```python
sns.lmplot(x ='smoothness_mean', y = 'compactness_mean',  
           data = df, hue = 'diagnosis') 
```

```python
X = np.array(df.iloc[:, 1:]) 
y = np.array(df['diagnosis'])
```

```python
from sklearn.model_selection import train_test_split 
X_train, X_test, y_train, y_test = train_test_split( 
    X, y, test_size = 0.33, random_state = 42) 
```

```python
from sklearn.impute import SimpleImputer

imputer = SimpleImputer(strategy='mean')
X_train = imputer.fit_transform(X_train)
X_test = imputer.transform(X_test)
```

```python
import numpy as np
non_nan_indices = ~np.isnan(X_train).any(axis=1)

X_train = X_train[non_nan_indices]
y_train = y_train[non_nan_indices] 
```

```python
import pandas as pd

X_train_df = pd.DataFrame(X_train)
y_train_df = pd.Series(y_train)

X_train_clean = X_train_df.dropna()
y_train_clean = y_train_df[X_train_clean.index]

X_train = X_train_clean.values
y_train = y_train_clean.values
```

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors=13) 
knn.fit(X_train, y_train)
```

```python
knn.score(X_test, y_test) 
```

```python
neighbors = [] 
cv_scores = [] 
  
from sklearn.model_selection import cross_val_score 
for k in range(1, 51, 2): 
    neighbors.append(k) 
    knn = KNeighborsClassifier(n_neighbors = k) 
    scores = cross_val_score( 
        knn, X_train, y_train, cv = 10, scoring = 'accuracy') 
    cv_scores.append(scores.mean()) 
```

```python
MSE = [1-x for x in cv_scores] 
  
optimal_k = neighbors[MSE.index(min(MSE))] 
print('The optimal number of neighbors is % d ' % optimal_k) 

plt.figure(figsize = (10, 6)) 
plt.plot(neighbors, MSE) 
plt.xlabel('Number of neighbors') 
plt.ylabel('Misclassification Error') 
plt.show() 
```
